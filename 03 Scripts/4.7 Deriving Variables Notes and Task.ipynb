## Creating a user-defined function for sorting price

```python
def price_label(row):

  if row['prices'] <= 5:
    
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range product'
  else: return np.nan
```

```python
#Using it on the dataframe
df_one_mil['price_range'] = df_one_mil.apply(price_label, axis=1)
```

```python
#Running the value counts function to check values in the new price_label column
df_one_mil['price_range'].value_counts(dropna = False)
```

```python
# Checking to see what the most expensive product in the subset is
df_one_mil['prices'].max()
```

## Summarizing how busy each day of the week is

```python
#Printing the frequency of the orders_day_of_the_week column
df_ords_products_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
 result = []

for value in df_ords_products_merged["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
# Creating a result_2 for the new column
result_2= []

for value in df_ords_products_merged["orders_day_of_week"]:
  if value == 0 and 1:
    result_2.append("Busiest days")
  elif value == 3 and 4:
    result_2.append("Slowest days")
  else:
    result_2.append("Regularly busy")
```
